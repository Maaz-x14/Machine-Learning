```python
import numpy as np
import matplotlib.pyplot as mplot
import matplotlib.image as mpi

arr = np.array(['fox.jpg','rabbit.jpg','bird.jpg','dear.jpg','snowwolf.jpg','whitewolf.jpg','owl.jpg','bear.jpg'])

for i in arr:
    img = mpi.imread(i)  # Read the image file
    mplot.imshow(img)  # Display the image
    mplot.axis('off')  # Hide the axis
    mplot.show()  # Show the image
```
