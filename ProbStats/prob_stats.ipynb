```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
```

```python
count = 0
size_arr = 100

# Discrete Random Var
rolls = np.random.choice([0, 1, 2, 3, 4, 5], size=size_arr) 

for i in rolls:
    if i == 0:
        count += 1

print("Ones: ", count)
print("Others: ", size_arr-count)

# Continuous Random Var
heights = np.random.normal(loc=170, scale=10, size=size_arr)  # mean=170, std=10
print("Dice: ", rolls+1)
print("Heights: ", heights)
```

```python
# Expectation (mean): E[X] = ∑ p(x)*x
# Variance: Var(X) = E[(X - μ)²]
# Covariance: how 2 variables move together.
# Correlation: normalized covariance, ranges [-1, 1].

x = np.random.normal(0, 1, 1000)
y = 2*x + np.random.normal(0, 1, 1000)

np.mean(x), np.var(x), np.cov(x, y)[0,1], np.corrcoef(x, y)[0,1]
```

```python
import random
import matplotlib.pyplot as plt

# Simulate 1000 dice rolls
roll_dice = np.random.choice([1,2,3,4,5,6], size=1000)
roll_dice

# Plot the distribution
plt.hist(roll_dice, bins=range(1, 8), align='left', edgecolor='black', rwidth=0.8)
plt.xticks(range(1, 7))
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.title('Distribution of 1000 Dice Rolls')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Generate two datasets
np.random.seed(42)
group1 = np.random.normal(loc=0, scale=1, size=1000)      # Standard normal
group2 = np.random.normal(loc=0.3, scale=1, size=1000)    # Slightly shifted mean

# Perform two-sample t-test
t_stat, p_value = ttest_ind(group1, group2)

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Result: Statistically significant difference (p < 0.05)")
else:
    print("Result: No significant difference")

# Optional: Plot distributions
plt.hist(group1, bins=50, alpha=0.6, label='Group 1 (mean=0)', color='blue')
plt.hist(group2, bins=50, alpha=0.6, label='Group 2 (mean=0.3)', color='purple')
plt.legend()
plt.title('Distribution Comparison')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
