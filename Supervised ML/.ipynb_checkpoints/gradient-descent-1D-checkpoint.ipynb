```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function: J(w) = (w - 3)^2 + 2
def J(w):
    return (w - 3)**2 + 2

# Derivative of J
def dJ_dw(w):
    return 2 * (w - 3)

# Gradient Descent
w = 10  # start far from min
alpha = 0.1
history = [w]

for i in range(20):
    w = w - alpha * dJ_dw(w)
    history.append(w)

# Plotting
w_vals = np.linspace(-2, 12, 100)
plt.plot(w_vals, J(w_vals), label="Cost Function J(w)")
plt.scatter(history, J(np.array(history)), c='red', label="Steps")
plt.xlabel("w")
plt.ylabel("J(w)")
plt.legend()
plt.show()
```
